# SQL - БД сервиса чтения книг по подписке

Необходимо проанализировать базу данных. В ней — информация о книгах, издательствах, авторах, а также пользовательские обзоры книг. Эти данные помогут сформулировать ценностное предложение для нового продукта.

**Задания**

- Посчитать, сколько книг вышло после 1 января 2000 года;

- Для каждой книги посчитать количество обзоров и среднюю оценку;

- Определить издательство, которое выпустило наибольшее число книг толще 50 страниц, т.е. исключить из анализа брошюры;

- Определить автора с самой высокой средней оценкой книг — учитывать только книги с 50 и более оценками;

- Посчитать среднее количество обзоров от пользователей, которые поставили больше 50 оценок.

**Описание данных**

- Таблица `books` Содержит данные о книгах:
 - book_id — идентификатор книги;
 - author_id — идентификатор автора;
 - title — название книги;
 - num_pages — количество страниц;
 - publication_date — дата публикации книги;
 - publisher_id — идентификатор издателя.

- Таблица `authors` Содержит данные об авторах:
 - author_id — идентификатор автора;
 - author — имя автора.

- Таблица `publishers` Содержит данные об издательствах:
 - publisher_id — идентификатор издательства;
 - publisher — название издательства;

- Таблица `ratings` Содержит данные о пользовательских оценках книг:
 - rating_id — идентификатор оценки;
 - book_id — идентификатор книги;
 - username — имя пользователя, оставившего оценку;
 - rating — оценка книги.
- Таблица `reviews` Содержит данные о пользовательских обзорах на книги:
 - review_id — идентификатор обзора;
 - book_id — идентификатор книги;
 - username — имя пользователя, написавшего обзор;
 - text — текст обзора.

```python
import pandas as pd
from sqlalchemy import create_engine

db_config = {'user': '***', # имя пользователя
            'pwd': '***', # пароль
            'host': '***',
            'port': 6432, # порт подключения
            'db': '***'} # название базы данных

connection_string = 'postgresql://{user}:{pwd}@{host}:{port}/{db}'.format(**db_config)

engine = create_engine(connection_string)
```

### Состав базы и типы данных в интересующих таблицах

```python
query0 = '''
SELECT *
FROM pg_catalog.pg_tables
WHERE schemaname != 'pg_catalog' AND
      schemaname != 'information_schema';
'''

display(pd.io.sql.read_sql(query0, con = engine))

query02 = '''
SELECT 
    table_name, 
    column_name, 
    data_type, 
    is_nullable
FROM INFORMATION_SCHEMA.COLUMNS 
WHERE table_name IN ('books', 'authors', 'publishers', 'ratings', 'reviews');
'''

display(pd.io.sql.read_sql(query02, con = engine))
```

### Первые строчки интересующих таблиц

```python
query1 = '''
             SELECT * FROM books
             LIMIT 5
        '''
query2 = '''
            SELECT * FROM authors
            LIMIT 5
        '''
query3 = '''
            SELECT * FROM publishers
            LIMIT 5
        '''
query4 = '''
            SELECT * FROM ratings
            LIMIT 5
        '''
query5 = '''
            SELECT * FROM reviews
            LIMIT 5
        '''

df_books = pd.io.sql.read_sql(query1, con=engine, index_col='book_id')
df_authors = pd.io.sql.read_sql(query2, con=engine, index_col='author_id')
df_publishers = pd.io.sql.read_sql(query3, con=engine, index_col='publisher_id')
df_ratings = pd.io.sql.read_sql(query4, con=engine, index_col='rating_id')
df_reviews = pd.io.sql.read_sql(query5, con=engine, index_col='review_id')

display(df_books)
display(df_authors)
display(df_publishers)
display(df_ratings)
display(df_reviews)
```

### Сколько книг вышло после 1 января 2000 года

```python
query6 = '''
            SELECT COUNT(book_id) AS Quantity_books_after_2000_1_1  FROM books
            WHERE CAST(publication_date AS date) >= '2000-01-01'
         '''

books_2000 = pd.io.sql.read_sql(query6, con=engine)
books_2000
```

Если сервис **для чтения книг по подписке**, то речь, собственно, про сами тексты, лишенные всех прочих органолептических свойств бумажных книг. В таком случае дата публикации совсем неважна. Например, новизна [Шекспира, выпущенного в 2012 году](#https://www.amazon.com/Hamlet-William-Shakespeare/dp/1451669410) как-то определяет его преимущество перед другими книгами? Вообще 1000 книг - это мало.

### Количество обзоров и средняя оценка для каждой книги

```python
query7 = '''
            WITH
             avg_rat AS 
                 (SELECT book_id, AVG(rating) AS avg_rating  FROM ratings
                 GROUP BY book_id),
             count_rew AS 
                 (SELECT book_id, COUNT(review_id) AS count_review  FROM reviews
                 GROUP BY book_id)
            SELECT c.title,
                   a.book_id,
                   a.avg_rating, 
                   b.count_review FROM avg_rat AS a 
            LEFT JOIN books AS c ON c.book_id=a.book_id
            LEFT JOIN count_rew AS b ON b.book_id=c.book_id
            ORDER BY a.avg_rating DESC
         '''

books_rating_review = pd.io.sql.read_sql(query7, con=engine, index_col='title')
display(books_rating_review.head(10))
books_rating_review.info()
```

Очень активные пользователи - не ленятся ставить рейтинги и писать обзоры. Только 6 книг вообще не имеют обзоров. Хотя, возможно проставление рейтинга является условием продолжения подписки, а написание обзора дает какие-то дополнительные бонусы участникам.

### Издательство, которое выпустило наибольшее число книг толще 50 страниц

```python
query8 = '''
            WITH
             books_fat AS 
                 (SELECT book_id, publisher_id FROM books
                 WHERE num_pages > 50)
            SELECT b.publisher,
                   COUNT(a.book_id) AS quantity_fat_books
            FROM books_fat AS a 
            LEFT JOIN publishers AS b ON b.publisher_id=a.publisher_id
            GROUP BY b.publisher           
            ORDER BY COUNT(a.book_id) DESC
         '''

books_publisher_fat = pd.io.sql.read_sql(query8, con=engine, index_col='publisher')
books_publisher_fat.head(10)
```

Никакой неожиданности - "Пингвин" с девизом `дешево и сердито` лидер беллетристики на все времена англоязычного мира (и среди бумажных книг тоже).

### Автор с самой высокой средней оценкой книг, учитываем только книги с 50 и более оценками

```python
query9 = '''
            WITH
             avg_rat AS 
                 (SELECT book_id, AVG(rating) AS avg_rating, COUNT(rating) FROM ratings
                  GROUP BY book_id
                  HAVING COUNT(rating) >= 50)
            SELECT b.author,
                   COUNT(a.book_id) AS books_in_rating,
                   AVG(a.avg_rating) AS avg_avg_rating
            FROM 
            avg_rat AS a LEFT JOIN
            books AS c ON c.book_id=a.book_id LEFT JOIN
            authors AS b ON b.author_id=c.author_id
            GROUP BY b.author
            ORDER BY AVG(a.avg_rating) DESC
         '''

books_author_rating = pd.io.sql.read_sql(query9, con=engine, index_col='author')
display(books_author_rating)
```

Дж.К.Роулинг - абсолютный победитель и в номинации кол-ва книг в рейтинге. Но как в рейтинг попали переводы на въетнамский язык? А соседство Шекспира с Паоло Коэльо (в двойном переводе на турецкий через английский) - это о чем говорит? Аудитория сервиса очень специфична? Куда катится мир?

### Среднее количество обзоров от пользователей, которые поставили больше 50 оценок

```python
query10 = '''
            WITH
             rating_spammers AS 
                 (SELECT username, COUNT(rating) AS count_ratings FROM ratings
                  GROUP BY username
                  HAVING COUNT(rating) > 50)
            SELECT AVG(count_reviews) AS avg_reviews_of_rating_spammers
            FROM (SELECT a.username,
                         COUNT(b.review_id) AS count_reviews,
                         AVG(a.count_ratings) AS count_ratings
                  FROM rating_spammers AS a 
                  LEFT JOIN reviews AS b ON b.username=a.username
                  GROUP BY a.username           
                  ORDER BY count_reviews DESC) AS reviws_spammers
         '''

books_reviewer_spam_avg = pd.io.sql.read_sql(query10, con=engine)
books_reviewer_spam_avg
```

В задаче не предполагается вывести доп. информацию о кол-ве оценок на пользователей и числе этих пользователей, хотя в промежуточных итогах это считается. Интересно, что больше 50-ти оценок оставили всего 6 спаммеров, т.е. обозревают каждую вторую книгу.
